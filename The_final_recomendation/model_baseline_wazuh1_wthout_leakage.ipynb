```python
import pandas as pd
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, f1_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

import matplotlib.pyplot as plt
```

```python
DATA_PATH = Path("/home/chupchik/Рабочий стол/fisrt_stepD/The_final_recomendation/output_dataset/wazuh_features.parquet")

df = pd.read_parquet(DATA_PATH)

print(df.shape)
df.head()
```

```python
target_col = "y_priority"

print(df[target_col].value_counts())
```

```python
text_features = ["full_log"]

numeric_features = [
    "hour", "dayofweek",
    "is_weekend", "is_night",
    "is_internal_ip",
    "has_400", "has_401", "has_403", "has_500",
    "has_sql", "has_admin", "has_wp"
]

categorical_features = [
    "program", "location", "scenario"
]

feature_cols = text_features + numeric_features + categorical_features

df["full_log"] = df["full_log"].fillna("")

X = df[feature_cols].copy()
y = df[target_col].copy()

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Используемые признаки:")
print(feature_cols)
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Классы:", label_encoder.classes_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Train:", X_train.shape, y_train.shape)
print("Test:", X_test.shape, y_test.shape)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "tfidf_full_log",
            TfidfVectorizer(max_features=3000, ngram_range=(1, 2)),
            "full_log"
        ),
        (
            "num",
            Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("scaler", StandardScaler())
            ]),
            numeric_features
        ),
        (
            "cat",
            Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("onehot", OneHotEncoder(handle_unknown="ignore"))
            ]),
            categorical_features
        )
    ],
    remainder="drop"
)
```

```python
model = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    random_state=42
)

pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", model)
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
print("Macro F1:", f1_score(y_test, y_pred, average="macro"))

print("\nClassification report:\n")
print(classification_report(
    y_test,
    y_pred,
    target_names=label_encoder.classes_,
    digits=4
))
```

```python
cm = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
disp.plot(ax=ax, xticks_rotation=45)
plt.title("Confusion Matrix — Logistic Regression Without Leakage")
plt.show()
```
